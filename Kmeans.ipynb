```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('turkiye-student-evaluation_generic.csv')
```

```python
df
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=2)
```

```python
kmeans.fit(df)
```

```python
kmeans.cluster_centers_
```

```python
labels=kmeans.labels_
```

```python
print(labels)
```

```python
labels.shape
```

```python
labels=pd.DataFrame(labels,columns=['cluster'])
labels.columns
```

```python
df=pd.concat([df,labels],axis=1)
```

```python
df
```

```python
df.drop('cluster',inplace=True,axis=1)
```
